```python
df = df[(df['plant_type'] == 'trees') & (df['plant_name'] == 'zelkova') & (df['plant_part'] == 'overall')]
```

```python
query = set(['slow growth', 'collapsed'])
```

```python
def jaccard_coef(s1, s2):
  return len(s1.intersection(s2)) / len(s1.union(s2))
```

```python
df.loc[:, 'coef'] = df['plant_damage'].apply(lambda s1: jaccard_coef(s1, query))
df = df.sort_values('coef', ascending = False)
```
